```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("GoldPrice.csv")
df
```

```python

# Convert 'Date' column to datetime format  
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Set 'Date' as the index  
df.set_index('Date', inplace=True)
```

```python
# Display the first few rows to check the changes  
print(df.head())
```

```python

# Display first and last five rows
print(df.head())
```

```python
print(df.tail())
```

```python
# Dataset info
df.info()
```

```python
# Checking for missing values
df.isnull()
```

```python
df.isnull().sum()
```

```python
# Summary statistics
print(df.describe())
```

```python
# Drop non-numeric columns before calculating correlation
df_numeric = df.select_dtypes(include=['number'])

```

```python
# Compute correlation matrix
correlation = df_numeric.corr()

```

```python
# Correlation values with Gold Price (GLD)
print("Correlation with Gold Prices:\n", correlation['GLD'])
```

```python
# Distribution of Gold Prices
sns.displot(df['GLD'], color='green', kde=True)
plt.title("Gold Price Distribution")
plt.xlabel("Gold Price")
plt.ylabel("Frequency")
plt.show()
```

```python


# Check available column names
print(df.columns)  # Make sure 'Gold Price' or any other correct name exists

# Plotting the gold price trend
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['GLD'], color='blue', linewidth=2)  # Use the correct column name
plt.xlabel("Year")
plt.ylabel("Gold Price")
plt.title("Gold Price Trends Over Time")
plt.grid(True)
plt.show()
```

```python
# Plot heatmap
plt.figure(figsize=(8, 8))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, cmap='Blues')
plt.title("Feature Correlation Heatmap")
plt.show()
```
